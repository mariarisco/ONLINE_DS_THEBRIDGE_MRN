### Ejercicio 0

Importa los paquetes y módulos que necesites a lo largo del notebook

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
plt.subplot(1,2,1) # primer subplot
plt.plot(x, np.sin(x))

plt.subplot(1,2,2) # primer subplot
plt.plot(x, np.cos(x));
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
fig, ax = plt.subplots(3,1,figsize=(6,12))

ax[0].plot(x, np.sin(x),color='red')
ax[1].plot(x, np.cos(x),color='green')
ax[2].plot(x, np.tan(x),color='blue')
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
# Crear la figura
fig = plt.figure(figsize=(8, 6))  # Crear figura con tamaño específico

# Crear subplots con add_subplot
ax1 = fig.add_subplot(2, 2, 1)  # Primer subplot
ax1.plot(x, np.sin(x))
ax1.set_title('Seno')

ax2 = fig.add_subplot(2, 2, 2)  # Segundo subplot
ax2.plot(x, np.cos(x))
ax2.set_title('Coseno')

ax3 = fig.add_subplot(2, 2, 3)  # Tercer subplot
ax3.plot(x, np.tan(x))
ax3.set_title('Tangente')

ax4 = fig.add_subplot(2, 2, 4)  # Cuarto subplot
ax4.plot(x, np.log(x+1))
ax4.set_title('Logaritmo')

fig.subplots_adjust(hspace=0.3, wspace=0.3);

```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
# Crear la figura
fig = plt.figure(figsize=(8, 8))  # Tamaño de la figura

# crear la primera figura
plt.subplot(2, 1, 1) 
plt.plot(x, np.cos(x),color='blue',label='coseno')
plt.plot(x,np.sqrt(x),color='orange',label='raíz cuadrada')
plt.legend();
```

```python
# Crear la figura
fig = plt.figure(figsize=(8, 8))  # Tamaño de la figura

# crear la primera figura
plt.subplot(2, 1, 1) 
plt.plot(x, np.cos(x),color='blue',label='coseno')
plt.plot(x,np.sqrt(x),color='orange',label='raíz cuadrada')
plt.title('Raíz cuadrada y coseno')
plt.legend()

# Crear subplot para la segunda fila
fig = plt.gcf()  # Obtener figura actual
ax2 = fig.add_subplot(2, 1, 2)  # Segunda fila, ocupando toda la fila

# Trazar en el segundo subplot
ax2.plot(x, np.sin(x), color='blue')
ax2.set_title('Seno');
```

Para terminar, reproduce la imagen de la figura

```python
plt.plot(x,np.sin(x),
        'o',
        markerfacecolor='#FFFF00',
        markeredgecolor='blue',
        markeredgewidth=1.5,
        ms=(8))
plt.title('Ejemplo de Marcadores')
```

```python
plt.scatter(x, np.sin(x),
            c='#FFFF00',         # Relleno amarillo brillante
            edgecolor='blue',    # Borde azul
            linewidths=1.5,      # Grosor del borde
            s=80)               # Tamaño del marcador (área en puntos^2)
plt.title('Ejemplo de Marcadores')
```
