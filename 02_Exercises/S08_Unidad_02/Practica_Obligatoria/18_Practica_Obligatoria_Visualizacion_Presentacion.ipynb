### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def scatter_plot_v2(df, col_num1, col_num2, col_cat=None, point_size=50, scale=1, show_legend=True):
    """
    Dibuja un diagrama de dispersin entre dos columnas numricas, opcionalmente coloreando y ajustando el tamao de los puntos.
    
    Args:
        df (pd.DataFrame): DataFrame que contiene los datos.
        col_num1 (str): Nombre de la primera columna numrica (eje X).
        col_num2 (str): Nombre de la segunda columna numrica (eje Y).
        col_cat (str, opcional): Nombre de la columna categrica para colorear. Default: None.
        point_size (int o str): Tamao fijo de los puntos o nombre de una columna que controla el tamao. Default: 50.
        scale (float): Factor para escalar los tamaos de los puntos si `point_size` es una columna. Default: 1.
        show_legend (bool): Controla si se muestra la leyenda. Default: True.
    """
    plt.figure(figsize=(10, 6))
    
    # Determinar el tamao de los puntos
    if isinstance(point_size, str):  # Si point_size es una columna
        if point_size in df.columns:
            sizes = df[point_size] * scale
        else:
            raise ValueError(f"La columna '{point_size}' no existe en el DataFrame.")
    else:  # Si point_size es un valor numrico
        sizes = point_size

    # Crear el grfico
    if col_cat:
        sns.scatterplot(
            data=df, x=col_num1, y=col_num2, hue=col_cat, size=sizes, sizes=(10, 200), palette='viridis', alpha=0.5
        )
        if show_legend:
            plt.legend(title=col_cat, bbox_to_anchor=(1.05, 1), loc='upper left')
    else:
        sns.scatterplot(
            data=df, x=col_num1, y=col_num2, size=sizes, sizes=(10, 200), color="blue", alpha=0.5
        )
        if show_legend:
            plt.legend(title="Tamao", bbox_to_anchor=(1.05, 1), loc='upper left')

    # Etiquetas y ttulo
    plt.title(f'Diagrama de dispersin: {col_num1} vs {col_num2}', fontsize=14)
    plt.xlabel(col_num1, fontsize=12)
    plt.ylabel(col_num2, fontsize=12)
    plt.tight_layout()

    # Mostrar el grfico
    plt.show()
    
```

```python
# Calcular las frecuencias relativas
relative_frequencies = df['alive'].value_counts(normalize=True).reset_index()
relative_frequencies.columns = ['alive', 'frequency']
relative_frequencies['frequency'] *= 100  # Convertir a porcentaje

# Crear el grfico con sns.barplot
plt.figure(figsize=(8, 6))
sns.barplot(
    data=relative_frequencies,
    x='alive',
    y='frequency',
    hue='alive',  # Usar hue para mapear la paleta
    dodge=False,  # Evitar separacin en barras al usar hue
    palette=['red', 'green'],  # Paleta personalizada
    alpha=0.7
)

# Aadir etiquetas y ttulo
plt.title('Frecuencia Relativa: Supervivencia en el Titanic', fontsize=16)
# plt.xlabel('Supervivencia (0 = No sobrevivi, 1 = Sobrevivi)', fontsize=12)
plt.xlabel("")
plt.ylabel('Porcentaje (%)', fontsize=12)

# Eliminar la leyenda ya que no es necesaria
plt.legend([], [], frameon=False)

# Mostrar porcentajes sobre las barras
for index, row in relative_frequencies.iterrows():
    plt.text(index, row['frequency'] + 1, f'{row["frequency"]:.1f}%', ha='center', fontsize=10)

# Personalizar el eje x para mostrar los valores originales
plt.xticks(ticks=[0, 1], labels=['No sobrevivi', 'Sobrevivi'])  # Etiquetas descriptivas

# Mostrar grfico
plt.tight_layout()
plt.show()
```

2- En esa ocasin los datos apoyan que se dio el "Las mujeres y los nios primero", pero...  


3- ... que la clase en la que se viajes tuvo una influencia significativa y podra explicar...  


4- ... las diferencias de tasas de supervivencia entre mujeres y nios.    


```python
# Calcular frecuencias relativas
df_relative = df.groupby(['alive', 'who']).size().reset_index(name='counts')
df_relative['relative_frequency'] = df_relative['counts'] / df_relative['counts'].sum() * 100

plt.figure(figsize=(8, 6))

# Graficar con frecuencias relativas
g = sns.catplot(
    x="alive",
    y="relative_frequency",
    hue="who",
    data=df_relative,
    kind="bar",
    palette="viridis",
    height=6,
    aspect=1.5
)

# Etiquetas y ttulo
g.set_axis_labels("Supervivencia", "Frecuencia Relativa (%)")
g.fig.suptitle("Frecuencias Relativas de Supervivencia por Quin", y=1.02)

plt.show()
```
