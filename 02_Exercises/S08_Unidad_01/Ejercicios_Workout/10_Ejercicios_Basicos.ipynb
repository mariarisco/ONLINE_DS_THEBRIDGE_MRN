### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
# Crear la figura
fig = plt.figure(figsize=(8, 6))  # Crear figura con tamao especfico

# Crear subplots con add_subplot
ax1 = fig.add_subplot(2, 2, 1)  # Primer subplot
ax1.plot(x, np.sin(x))
ax1.set_title('Seno')

ax2 = fig.add_subplot(2, 2, 2)  # Segundo subplot
ax2.plot(x, np.cos(x))
ax2.set_title('Coseno')

ax3 = fig.add_subplot(2, 2, 3)  # Tercer subplot
ax3.plot(x, np.tan(x))
ax3.set_title('Tangente')

ax4 = fig.add_subplot(2, 2, 4)  # Cuarto subplot
ax4.plot(x, np.log(x+1))
ax4.set_title('Logaritmo')

fig.subplots_adjust(hspace=0.3, wspace=0.3);

```

```python
# Crear la figura
fig = plt.figure(figsize=(8, 8))  # Tamao de la figura

# crear la primera figura
plt.subplot(2, 1, 1) 
plt.plot(x, np.cos(x),color='blue',label='coseno')
plt.plot(x,np.sqrt(x),color='orange',label='raz cuadrada')
plt.title('Raz cuadrada y coseno')
plt.legend()

# Crear subplot para la segunda fila
fig = plt.gcf()  # Obtener figura actual
ax2 = fig.add_subplot(2, 1, 2)  # Segunda fila, ocupando toda la fila

# Trazar en el segundo subplot
ax2.plot(x, np.sin(x), color='blue')
ax2.set_title('Seno');
```
