# Visualizacin Aplicada: Mostrando una Variable Categrica

Nos centramos ahora en visualizar una nica variable (tanto para anlisis como presentacin). Dividimos en dos grupos categricas y nmericas. En esta sesin trataremos las categricas.

Lo primero, importamos libreras y cargamos datos:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
df_titanic = pd.read_csv("./data/titanic.csv")
df_seguros = pd.read_csv("./data/Marketing-Customer-Analysis.csv")
```

### Grficas y consideraciones

Veremos las siguientes:
- Diagrama de barras para frecuencias
- Diagrama de esferas/crculos para frecuencias
- Quesos, donuts y otras cosas de comer: lolipops


#### Consideraciones generales: 

* **Cundo**: Anlisis univariante de frecuencias y presentacin de distribucin de valores cuando esto aporte algo (en general para introducciones y dar contexto). Los diagramas de tarta y queso funcionan mejor con frecuencias relativas que con absolutas. Los lolipops no son de mi gusto, pero para que tengas otros.

* **Cundo no**: Incluso en la situacin de que sea necesario por dar contexto hablar de los valores de una variable categrica, si la cardinalidad es mayor de 5 (ms o menos) reducirla mostrando 4-5 valores como mucho colapsando los no interesantes en un grupo "otros" (ojo, los no interesantes no son los de menor frecuencia son los que no aporten nada a los mensajes a transmitir )

### Barras

```python
### Creamos figura y axes
fig,axs = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
fig.suptitle("Barras en Seaborn");
```

```python
### AC
### Countplot, nos permite frecuencias

# primer subplot, hue para asignar un color diferente a cada valor de la categora
sns.countplot(x = "embark_town", data = df_titanic, ax = axs[0], hue = "embark_town")
# Configurar el ttulo del subplot
axs[0].set_title("Frecuencias Absolutas de embark_town")

fig
```

```python
# Para frecuencias relativas... tambin hay que calcularlas previamente y puedes usar el barplot (si parecido a matplotlib)
valores = df_titanic.embark_town.value_counts(normalize = True) * 100
sns.barplot(x = valores.index, y = valores.values, hue = valores.index, ax = axs[1])
axs[1].set_title("Frecuencias relativas embark_town")
axs[1].set_ylabel("%")
fig
```

### Circulos

```python
### Creamos figura y axes
fig, axs = plt.subplots(nrows=1,ncols=1,figsize=(14,4))
fig.suptitle("Circulos en Seaborn");
```

```python
### Frecuencias absolutas y relativas
axs.cla

frecuencias = df_seguros["state"].value_counts() # frecuencia absoluto
df_frecuencias = df_seguros["state"].value_counts().reset_index() # primera columna con los valores de los estados y en la segunda el resultado del value counts
df_frecuencias.columns = ["categorias","frecuencias"]


sns.scatterplot(x = "categorias", y = [1]*len(frecuencias), hue = "categorias", data = df_frecuencias, size = "frecuencias", legend = False, ax = axs, sizes = (500,5000))
axs.set_xlabel("")


# aadir texto
for estado,valor in frecuencias.items(): # recorro la serie (como un diccionario)
    axs.text(estado,1.02,f"{valor}({round(valor*100/frecuencias.sum())}%)")
axs.set_facecolor("none") #quitar la cuadrcula de fondo
axs.yaxis.set_ticks([]) # para quitar los ticks de los ejes

fig
```

### Donuts, tartas, quesos

Seaborn no viene con ellos, as que usamos matplotlib, empezamos con la tarta/queso:

```python
data = df_titanic["class"].value_counts() # serie con las frecuencias

fig,ax = plt.subplots(1,1,figsize = (4,4))

ax.pie(data.values, # con matplotlib
        labels=data.index,
        autopct='%.2f%%', startangle= 90);
```

Ahora tipo donut:

```python
my_circle = plt.Circle( (0,0),
                       0.5, # grosor del donut
                       color = "white")

ax.add_artist(my_circle)

fig
```

### Lolipops

Este tampoco lo tenemos en seaborn:

```python
conteo = df_seguros['vehicle_class'].value_counts(ascending=False) #serie con las frecuencias

plt.figure(figsize=(10,5))

plt.hlines(y=conteo.index,
           xmin= 50,
           xmax=conteo,
           color='skyblue')

p=plt.gcf()
p.gca().set_facecolor("none") #borrar el fondo

plt.plot(conteo, conteo.index, "o");
```
